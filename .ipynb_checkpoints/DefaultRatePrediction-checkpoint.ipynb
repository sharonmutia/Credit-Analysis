# 6. Feature Selection

### a.) SelectKMeans
The Financial Measure 2 metrics and the average amount difference were disregarded from the initial chi2 test since the test assumes that the variables are distributions and therefore non-negative. The features were later re-introduced into the model and the model was evaluated to see if the improved or worsed the model.
I used the chi-square test of significance since it tests for independence/dependence between the y variable and the X variables.

The model with Average Amount Difference in addition to the chi-square parameters yeilded the highest precision results for the defaulters thus the variables were used in estimation

```python
categorical_data=['CustomerType', 'SOR', 'LoanCategory', 'CustomerLevel', 'DisbursementDate_month', 'MonthTime']
for column in categorical_data:
    df[column]=df[column].astype(int,inplace=True)
```

```python
df.min()
```

```python
#Best precision model
X=df[[ 'Age', 'CustomerType', 'SOR',
       'MonthsSinceOpen', 'CRBScore',
       'LoanCategory', 'DisbursementDate_day', 'DisbursementDate_month',
       'DefaultRate', 'FinancialMeasure1AveHist',
       'FinancialMeasure3AveHist', 'AmountAveHist']]
y=df['Default']

XTrain, XTest, yTrain, yTest = train_test_split(X, y, test_size=0.2, random_state=0, stratify=y)
```

```python
SelectFeatures=SelectKBest(chi2,k=5).fit(XTrain,yTrain)
```

```python
SelectedFeatures=pd.DataFrame({'Feature':['Age', 'CustomerType', 'SOR',
       'MonthsSinceOpen', 'CRBScore',
       'LoanCategory', 'DisbursementDate_day', 'DisbursementDate_month',
       'DefaultRate', 'FinancialMeasure1AveHist',
       'FinancialMeasure3AveHist', 'AmountAveHist'],'Ranking':SelectFeatures.scores_})
SelectedFeatures.sort_values(by='Ranking',ascending=False)
```

```python
XTrainChi2=SelectFeatures.transform(XTrain)
XTestChi2=SelectFeatures.transform(XTest)
```

```python
XTrainChi2.shape
```

### Original Model

```python
RandomForest = RandomForestClassifier(n_estimators=100, random_state=42)
RandomForest.fit(XTrain,yTrain)

yPred = RandomForest.predict(XTest)

print( "The model's accuracy is ", metrics.accuracy_score(yTest, yPred)*100)
metrics.accuracy_score(yTest, yPred)
print(metrics.classification_report(yTest, yPred))
```

### Chi-Square Model
